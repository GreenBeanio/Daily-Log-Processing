```python
from src.parser import parser, expressions, modes
```

Lexer Testing

```python
at_at_mode = tokens.LexerToken(search_term = "@@", lexer_token = tokens.LexerTokenEnum.TIME)
at_less_mode = tokens.LexerToken(search_term = "@<", lexer_token = tokens.LexerTokenEnum.COMMENT)
greater_at_mode = tokens.LexerToken(search_term = ">@", lexer_token = tokens.LexerTokenEnum.ACTIVITY)

lexer_tokens = tokens.LexerTokens()
lexer_tokens.addToken(at_at_mode)
lexer_tokens.addToken(at_less_mode)
lexer_tokens.addToken(greater_at_mode)

test_lexer = lexer.Lexer(lexer_tokens)
```

Parser Testing

```python
time_exp = expressions.ParserExpression(
    [tokens.LexerTokenEnum.SENTENCE,
     tokens.LexerTokenEnum.TIME],
     modes.ParserTimeMode())
activity_exp = expressions.ParserExpression(
    [tokens.LexerTokenEnum.SENTENCE,
     tokens.LexerTokenEnum.ACTIVITY],
     modes.ParserSentenceMode())
activity_comment_exp = expressions.ParserExpression(
    [tokens.LexerTokenEnum.SENTENCE, 
     tokens.LexerTokenEnum.COMMENT,
     tokens.LexerTokenEnum.SENTENCE, 
     tokens.LexerTokenEnum.ACTIVITY],
    modes.ParserActivityCommentMode())
comment_exp = expressions.ParserExpression(
    [tokens.LexerTokenEnum.COMMENT,
     tokens.LexerTokenEnum.SENTENCE,
     tokens.LexerTokenEnum.END_OF_INPUT],
     modes.ParserSentenceMode())
end_of_input_exp = expressions.ParserExpression(
    [tokens.LexerTokenEnum.END_OF_INPUT],
     modes.ParserEOIMode())

parser_exps = expressions.ParserExpressions()
parser_exps.addExp(time_exp)
parser_exps.addExp(activity_exp)
parser_exps.addExp(activity_comment_exp)
parser_exps.addExp(comment_exp)
parser_exps.addExp(end_of_input_exp)
```
